# Data visualisation
This notebook is for visualising some of the incomre-related data from CDC.gov
The data sets can be found here: https://www.cdc.gov/nchs/nhis/shs/tables.htm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### function to plot trends

```python
def plot_trends(fname = '2013_SHS_Sadness.csv', plt_title='Trend w.r.t income'):
    csv_dtypes = {'Income (USD)':str}
    #change path to datasets/cdc_data later
    df = pd.read_csv('../datasets/cdc_data/'+fname, encoding='cp1252', dtype=csv_dtypes)
    col_idx = df.columns
    plot_col = col_idx.tolist()
    print(plot_col)
    plot_col.remove('Income (USD)')
    df.plot(x='Income (USD)', y=plot_col, kind='bar', figsize=(15, 5), 
            fontsize=16, rot=1, 
            title = plt_title,
            colormap='plasma')
```

## Self assessment of health

```python
plot_trends(fname='2013_SHS_self_assessment.csv', plt_title='Self report of health by income')
```

## Cancer data

```python
plot_trends(fname='2013_SHS_Cancer.csv', plt_title='Cancer statistics by income')
```

## Mental health
This shows feelings of sadness, hoplessness etc. w.r.t income

```python
plot_trends(fname='2013_SHS_Sadness.csv', plt_title='Feelings of sadness, hopelessness, worthlessness all or most of the time')
```

## General diseases

```python
plot_trends(fname='2013_SHS_general_diseases.csv', plt_title='Common diseases by income')
```
